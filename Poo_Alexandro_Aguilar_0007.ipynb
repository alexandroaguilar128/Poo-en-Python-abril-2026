#### Conceptos de POO (Programacion orientada a objetos) en Python

### El prompt en IA Gemini
Lenguaje Python, Crear la clase "animal" con los atributos "nombre", "Raza" crear el constructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos.

```python
# Definimos la clase "animal"
class animal:
    # El constructor __init__ inicializa los atributos al crear el objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función (método) para ladrar
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau! 🐱‍👤")

    # Función (método) para correr
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente. 🐱‍🏍")

    def Morder(self):
        print(f"{self.nombre} está mordiendo a la vecina 🤦‍♀️")

# --- Creación de objetos (Instancias) ---

# Creamos el objeto 'pitbull' con datos específicos
pitbull = animal("Thor", "Pitbull")

# Creamos el objeto 'chihuahua' con datos específicos
chihuahua = animal("Copito", "Chihuahua")
# creamos el objeto perro callejero con datos
Perrocallejero = animal("Solovino", "Chafa")


# --- Uso de los objetos ---

print(f"Datos del primer perro: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()

print("-" * 30)

print(f"Datos del segundo perro: {chihuahua.nombre}, Raza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()
print("-" * 30)

print(f"Datos del tercer perro: {Perrocallejero.nombre}, Raza: {Perrocallejero.raza}")
Perrocallejero.ladrar()
Perrocallejero.correr()
Perrocallejero.Morder()
print(" Programa creado por Alexandro Aguilar NL = 1")
```
